```python
def create_matrix_for_apriori(df, matrix):
    length = len(df)
    width =len(df.columns)
    for i in range(0,length):
#         attributes = df_bbl_in_ha.iloc[i,:].apply(str).values
#         row = ",".join(attributes)
        matrix.append([str(df.values[i,j]) for j in range(0,width)])
    return matrix
        
```

```python
def format_output(output):
    res = []
    for record in output:
        frozset = record[0]
        support = record[1]
        ordered_statistics = record[2]
        res.append([frozset,support,ordered_statistics])
    res = pd.DataFrame(res, columns=['frozset','support','ordered_statistics'])
    return res
```

```python
prep_metrix = []
prep_metrix = create_matrix_for_apriori(df_bbl_in_ha,prep_metrix)
```

```python
prep_metrix
```

```python
from apyori import apriori
rule = apriori(transactions = prep_metrix, min_support = 0.6, min_confidence = 0.8)
output_ha=list(rule)
```

```python

res_ha =format_output(output_ha)
res_ha.sort_values('support', ascending = False)
```

```python
res_ha.to_csv("./nycha_output.csv")
```
