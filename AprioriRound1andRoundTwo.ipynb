```python
import csv
import sys
csvMatrix = []
# "bldgclass","landuse","lotarea", "numbldgs","histdist"
with open("./small.csv", 'r') as csvfile:
    myreader = csv.reader(csvfile)
    next(myreader)
    for row in myreader:
        csvMatrix.append(row)
```

```python
csvMatrix
```

```python
rule = apriori(transactions = csvMatrix, min_support = 0.5, min_confidence = 0.6)
```

```python
output = list(rule)
len(output)
```

```python
output[0]
```

```python
res = []

for record in output:
    frozset = record[0]  # some formula for obtaining values
    support = record[1]
    ordered_statistics = record[2]
    res.append([frozset,support,ordered_statistics])

df = pd.DataFrame(res, columns=['frozset','support','ordered_statistics'])
print(df)
```

```python
df.to_csv("./output.csv")
```

```python
def groupItems(df):
    ls = df["frozset"].apply(lambda x: ', '.join(list(x))).astype("unicode").tolist()
    item=[]
    itemCount= dict()
    for l in ls:
        item.extend(l.split(', '))
    for r in item:
        if r in itemCount:
            itemCount[r] = itemCount[r] +1
        else:
            itemCount[r] = 1
    return itemCount
```

```python
itemCount =groupItems(df)
itemCount
```
